```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
import math
import seaborn
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 10)
seaborn.set_context("talk")
```

```python
from scipy.stats import multivariate_normal
```

```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
def show(data):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(data[:, 0], data[:, 1], data[:, 2])
```

```python
def countour(data):
    d = data[:, 2]
    CS = plt.contour(np.linspace(-3,3), np.linspace(-3,3),  d.reshape(math.sqrt(d.shape[0]), -1 ).transpose())
```

```python
def compute(mu, sigma):
    var = multivariate_normal(mean=mu, cov=sigma)
    data = [(x, y, var.pdf([x,y])) for x in np.linspace(-3, 3) for y in np.linspace(-3, 3)]
    return np.array(data)
```

```python
show(compute([0,0], [[1,1.5],
                     [1,3]]))
```

```python
countour(compute([0,0], [[1,1.5],
                         [1, 3]]))
```

```python
#plt.ylim([-3,5])
#plt.xlim([-3,5])
countour(compute([0.,0.], [[2.,0.],
                         [0., 2.]]))
```
